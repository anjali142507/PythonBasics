```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = load_breast_cancer()
x = data.data
y = data.target
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2)
```

```python
pipeline = Pipeline([
    ('scaler' , StandardScaler()),
    ('svm' , SVC())
])
```

```python
parameters = {
    'svm__C' : [0.1,1,10],
    'svm__kernel':['linear' , 'rbf'],
}
```

```python
grid_search = GridSearchCV(pipeline , parameters, cv = 5)
grid_search.fit(x_train , y_train)
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
classification_rep = classification_report(y_test,y_pred)
print(classification_rep)
```

```python
cross_val_scores = cross_val_score(best_model ,x, y, cv = 5)
mean = cross_val_scores.mean()
print(mean)
```

```python
import gym
import torch
import torch.nn as nn
import torch.optim as open
import torch.optim as optim
env=gym.make('CartPole-v1')
```

```python
model = nn.Sequential(
    nn.Linear(env.observation_space.shape[0],64),
    nn.ReLU(),
    nn.Linear(64,64),
    nn.ReLU(),
    nn.Linear(64, env.action_space.n)
)
```

```python
optimizer = optim.Adam(model.parameters(),lr=0.001)
criterion = nn.CrossEntropyLoss()
```

```python
n=100
for episode in range(n):
  state = env.reset()
  done = False
  tr=0
  while not done:
    action_probs = torch.softmax(model(torch.tensor(state,dtype = torch.float32)),dim =-1)
    action = torch.multinomial(action_probs,num_samples = 1).item()
    next_state, reward,done,_=env.step(action)
    optimizer.zero_grad()
    loss = criterion(model(torch.tensor(state,dtype=torch.float32)),torch.tensor(action))
    loss.backward()
    optimizer.step()
    state = next_state
    tr+=reward
    print("episode",episode+1,"Reward=",tr)

  env.close()

```
